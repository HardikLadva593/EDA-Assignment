```python
app_categorical = app_new.select_dtypes(include=['object']).columns
app_categorical
```

```python
def plotting(app_new, app0, app1, column):
    
    app_new = app_new
    app_0 = app0
    app_1 = app1
    col = column
    
    fig = plt.figure(figsize=(13,10))
    
    ax1 = plt.subplot(221)
    app_new[col].value_counts().plot.pie(autopct = "%1.0f%%", ax=ax1)
    plt.title('Plotting data for the column: '+ column)
    
    ax2 = plt.subplot(222)
    sns.countplot(x= column, hue = 'TARGET', data = app_new, ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Plotting data for target in terms of total count')


    ax3 = plt.subplot(223)
    app_new = pd.DataFrame()
    app_new['0']= ((app_0[col].value_counts())/len(app_0))
    app_new['1']= ((app_1[col].value_counts())/len(app_1))
    app_new.plot.bar(ax=ax3)
    plt.title('Plotting data for target in terms of percentage')


    fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

    plt.show()
      
```

```python
for column in app_categorical:
    print("Plotting ",column)
    plotting(app_new, app_0, app_1, column)
    print("=======================================================")
```
